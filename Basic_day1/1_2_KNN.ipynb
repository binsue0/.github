# K-Nearest Neighbor

https://github.com/songsuoyuan/The-Element-of-Statistical-Learning-Python-Implements

```python
import numpy as np
from sklearn import neighbors
import matplotlib.pyplot as plt
```

```python
data_path_url = 'https://raw.githubusercontent.com/binsue0/.github/main/Basic_day1/data/'
data = np.loadtxt(data_path_url + 'mixture_simulation_data.txt', skiprows=1)
means = np.loadtxt(data_path_url + 'means.txt')
X = data[:,:2]
y = data[:,2]
```

```python
def k_nearest_neighbor(X, y, k, p=2):
    '''
    K nearest neighbor method:
        y(x) = 1 / k * (y_1 + ... + y_k)
    where y_i belongs to the k closest points to the point x
    Complexity: based on implement, brute force is O(Dn^2)

    sklearn.neighbors.KNeighborsClassifier : p(float, default=2)
    Power parameter for the Minkowski metric.
    When p = 1, this is equivalent to using manhattan_distance (l1), and euclidean_distance (l2) for p = 2. For arbitrary p, minkowski_distance (l_p) is used.
    '''

    # 1) 모델 학습
    clf = neighbors.KNeighborsClassifier(n_neighbors=k, p=p)
    clf.fit(X, y)

    # 2) 결정 경계(decision boundary)를 그리기 위한 예측 격자 생성
    δ = 0.1
    xs = np.arange(X[:,0].min()-0.5, X[:,0].max()+0.5, δ)
    ys = np.arange(X[:,1].min()-0.5, X[:,1].max()+0.5, δ)
    GX, GY = np.meshgrid(xs, ys)
    pts = np.c_[GX.ravel(), GY.ravel()]

    # 3) 격자 예측 및 형태 복원
    Z = clf.predict(pts).reshape(GX.shape)

    # 4) 시각화
    plt.figure(figsize=(8,8))
    plt.scatter(X[:,0], X[:,1], c=y, alpha=0.6)      # 데이터 산점도
    plt.contour(GX, GY, Z, levels=[Z.mean()],        # 결정 경계
                alpha=0.8, colors='b', linewidths=3)
    plt.show()

    # 5) 정확도 계산
    preds = np.round(clf.predict(X))
    acc = (preds == y).mean() * 100
    print(f'Precision: {acc:.2f}%')
```

```python
k_nearest_neighbor(X, y, 15, p=2) # Example with p=2 for euclidian distance
```

```python
k_nearest_neighbor(X, y, 15, p=1)  # Example with p=1 for Manhattan distance
```
